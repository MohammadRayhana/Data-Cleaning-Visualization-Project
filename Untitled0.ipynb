```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("quikr_car.csv")

# Remove duplicates
df.drop_duplicates(inplace=True)

# Handle missing values
df["fuel_type"].fillna("Unknown", inplace=True)

# Remove rows with missing kms_driven
df = df[df["kms_driven"].notna()]

# Convert kms_driven
df["kms_driven"] = (
    df["kms_driven"]
    .str.replace(" kms","",regex=False)
    .str.replace(",","",regex=False)
)

df = df[df["kms_driven"].str.isnumeric()]
df["kms_driven"] = df["kms_driven"].astype(int)

# Clean Price column
df = df[df["Price"] != "Ask For Price"]

df["Price"] = (
    df["Price"]
    .astype(str)
    .str.replace(",","",regex=False)
)

df = df[df["Price"].str.isnumeric()]
df["Price"] = df["Price"].astype(int)

# Clean Year column
df = df[df["year"].astype(str).str.isnumeric()]
df["year"] = df["year"].astype(int)

print(df.info())
```

```python
Q1 = df["Price"].quantile(0.25)
Q3 = df["Price"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df["Price"] >= lower) &
        (df["Price"] <= upper)]
```

```python
import matplotlib.pyplot as plt

top_brands = df["company"].value_counts().head(10)

top_brands.plot(kind="bar")

plt.title("Top 10 Car Brands")
plt.xlabel("Brand")
plt.ylabel("Number of Cars")
plt.show()
```

```python
df["fuel_type"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Fuel Type Distribution")
plt.ylabel("")
plt.show()
```

```python
import seaborn as sns

sns.histplot(df["Price"], bins=30)

plt.title("Car Price Distribution")
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x="kms_driven",
    y="Price"
)

plt.title("Price vs Kilometers Driven")
plt.show()
```

```python
avg_price = (
    df.groupby("company")["Price"]
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

avg_price.plot(kind="bar")

plt.title("Top Brands by Average Price")
plt.show()
```
